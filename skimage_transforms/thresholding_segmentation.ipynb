```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```


# thresholding ---- intensities greater and lesser than a threshold is converted to particular intensity value ----
used for segmentation----
first is to convert to grayscale

```python

img = cv2.imread('tree.jpg',0)
print(img.shape)
plt.imshow(img,cmap='gray')
plt.imsave('grayimage.jpeg',arr = img)
```

```python
print("the maximum pixel value is",img.max())
img
```

```python

threshold = 127
binary = img > threshold
plt.imshow(binary)
plt.title("thresholding for segmentation")
plt.show()
```

```python
binary_invert = img <= threshold
plt.imshow(binary_invert)
plt.title("inverted thresholding for segmentation")
plt.show()
```

# --- all possible thresholding using scikit image ---- using try_all _thresholding function in filters module ---

```python
from skimage.filters import try_all_threshold
fig, ax = try_all_threshold(img,verbose=False)
#shows the seven thresholding algorithms used
```

thresholding --- otsu method --- threshold can be calculated from the intensity values of the images itself

```python
from skimage.filters import threshold_otsu
threshold = threshold_otsu(img)
binary1 = img > threshold
plt.imshow(binary1)
plt.title("thresholding for segmentation using otsu method")
plt.show()
```
