# filtering --- neighbourhood operation --- sobel filter for edge detection
smoothing,edge detection,enhancing,sharpening

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from skimage.filters import sobel
import cv2
```

```python
gray_imag = cv2.imread("grayimage.jpeg",0)
gray_imag.shape
```

```python
edge_sobel = sobel(gray_imag)
plt.imshow(edge_sobel)
plt.title("plot showing the edges using sobel filter ")
plt.show()
```

# contrast stretching
low contrast image have their histogram (distribution of intensity values) in mid values and small range
so we do contrast stretching or histogram equalizer

```python
plt.hist(gray_imag)
plt.show()
```

```python
from skimage import exposure
equalised_img = exposure.equalize_hist(gray_imag)
plt.imshow(equalised_img,cmap='gray')
```

```python
gray_imag1 = cv2.imread("lungs.jfif",0)
plt.imshow(gray_imag1,cmap='gray')
plt.show()
equalised_img = exposure.equalize_hist(gray_imag1)
plt.imshow(equalised_img,cmap='gray')
plt.show()
```

# tranformation 
compressing,resizing,optimising,rotation...

```python
from skimage.transform import rotate
```

```python
rotated_imag = rotate(gray_imag,90)
plt.imshow(rotated_imag,cmap='gray')
```

```python
from skimage.transform import rescale
#rescaling the image to 4 times smaller
rescaled_img = rescale(gray_imag1,1/4,anti_aliasing=True,multichannel=False)

#when aliasing is true the downscaled image is smoothed by a filter for better visualisation
#multichannel is made true to indicate a colored image

plt.imshow(rescaled_img,cmap='gray')
plt.show()
```

```python
from skimage.transform import resize
#resize to a fixed number i.e height and width without any proportion
height = 300
width = 500
resized_img = resize(gray_imag,(height,width),anti_aliasing=True)
plt.imshow(resized_img)
plt.show()
```
